```python
import pandas as pd
import matplotlib.pyplot as plt

# List of metrics file names
metrics_dir = "train_metrics"
# metrics_dir = "notebook/train_metrics_bak"
metrics_files = ["case3.csv",  "case22.csv", "case27.csv",  "case13.csv", "case20.csv", "case26.csv"]

# Define labels as they appear in the legend
labels = ["o-4","m-4",  "c-4", "o-5",  "m-5", "c-5"]

# Plot each file
for metrics_file, label in zip(metrics_files, labels):
    data = pd.read_csv(f"{metrics_dir}/{metrics_file}")
    epochs = data["epoch"]
    auprc_values = data["val_auc_prc"]
    plt.plot(epochs, auprc_values, label=label)

# Set plot labels and title
plt.xlabel("Epoch")
plt.ylabel("AUPRC")
plt.legend()
plt.grid(True)

# Zoom in on the AUPRC values from 0.8 to 1
plt.ylim(0.9, 0.96)

# Show plot
plt.show()
```
